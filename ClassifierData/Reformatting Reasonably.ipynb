```python
import pandas as pd
import os
import numpy as np

root = os.getcwd()
df = pd.read_csv(os.path.join(root, "dataset_cleaved_uncleaved.csv"), header = None)

# assuming these one-hot vectors are in order of printout ... there is no other way to make a guess
num_amino = 20
amino_order = "ACDEFGHIKLMNPQRSTVWY"
```

```python
df.head()
```

```python
# newdf is only the one hot vectors to ensure nothing we dont expect occurs

newdf = pd.DataFrame(np.zeros(shape = (len(df.iloc[:,0]), 100)))
for i in range(100):
    newdf.iloc[:,i] = df.iloc[:, i]
newdf.head()
```

```python
# final_df will be a nice txt file with two columns: sequence, result

final_df = pd.DataFrame(np.zeros(shape = (len(df.iloc[:,0]), 2)))
for i in range(len(df.iloc[:,0])):
    res = df.iloc[i, 104] # cleaved or uncleaved
    seq = ""
    for j in range(5):
        one_hot = list(df.iloc[i,(20*j):(20*(j + 1))])
        index = one_hot.index(1)
        aa = amino_order[index]
        seq += aa
    final_df.iloc[i, 0] = seq
    final_df.iloc[i, 1] = res
final_df.columns = ["Sequence", "Result"]
final_df["Sequence"] = final_df["Sequence"].map(lambda x: "A{}C.ASHL".format(x))
final_df.head()
```

```python
final_df.to_csv("experimental_binary_classifications.txt", index = None, sep = "\t")
```

```python
fh = open("HCV_sequences.txt", "w")
for el in final_df["Sequence"]:
    fh.write(el + "\n")
fh.close()
```
