```python
def visualize_A(A):
    """takes in A in the format (NxNxM) for visualization"""
    fig, axs = plt.subplots(A.shape[2], 1, figsize=(5, 15), sharey=True)
    for m in range(A.shape[2]):
        axs[m].imshow(A[:,:,m], cmap='hot', aspect='auto')
        plt.title("A Visualized")
        plt.xlabel("Residues")
        plt.ylabel("Residues")
        plt.axis('off')
    plt.show()
    
def visualize_V(V):
    """takes in V in the format (NxK) for visualization"""
    fig, ax = plt.subplots()
    ax.imshow(V, cmap='hot', aspect='auto')
    plt.title("V Visualized")
    plt.xlabel("Feature")
    plt.ylabel("Residues")
    plt.show()
```

```python
import numpy as np
import pickle as pkl
import sys
import os
import scipy.sparse as sp


def graph_list_pickle(graph_ls, label_ls, dataset_name, destination_path):
    """Takes in a list of graphs and labels and pickles them in proper format. It also puts an index file in the directory"""
    
    # get the root
    root = os.getcwd()
    try:
        os.chdir(destination_path)
        os.chdir(root)
    except:
        raise ValueError("You're destination folder does not exist")
    
    # get population size, number of nodes, and number of node features
    population = len(graph_ls)
    N = graph_ls[0].V.shape[0]
    F = graph_ls[0].V.shape[1]
    
    # find number of classifications possible
    s = set()
    for el in label_ls:
        s.add(el)
    s = list(s)
    num_classifiers = len(s)
    print(s)
    
    # generate feature matrices
    x = np.zeros(shape = (population, N, F))
    y = np.zeros(shape = (population, num_classifiers), dtype = np.int64)
    graph = np.zeros(shape = (population, N, N))
    
    print("num classifiers{}".format(num_classifiers))
    
    # populate all elements
    for i in range(population):
        x[i,:,:] = graph_ls[i].V
        graph[i,:,:] = graph_ls[i].A[:,:,0] # the third dim is to get first adj matrix
        if np.isnan(x[i,:,:]).any():
            print("ya done fucked up")
            print(np.argwhere(np.isnan(x[i,:,:])))
            visualize_V(x[i,:,:])
            print(x[i,:,:].shape)
        else:
            print("no nan detected")
        one_hot_label_vec = np.zeros(num_classifiers)
        one_hot_label_vec[s.index(label_ls[i])] = 1 
        y[i,:] = one_hot_label_vec
    
    # get indices of the test set
    idx = [x for x in range(population)]
    np.random.shuffle(idx)
    test_fraction = .3
    cutoff = int(len(idx) * test_fraction)
    test_index = idx[:cutoff]
    
    # pickle everything
    pkl.dump(x, open( os.path.join(destination_path,\
            "ind.{}.x".format(dataset_name)), "wb"))
    pkl.dump(y, open( os.path.join(destination_path,\
            "ind.{}.y".format(dataset_name)), "wb"))
    pkl.dump(graph, open( os.path.join(destination_path,\
            "ind.{}.graph".format(dataset_name)), "wb"))
    
    f = open( os.path.join(destination_path, "ind.{}.test.index".format(dataset_name)), "w")
    for el in test_index:
        f.write(str(el) + "\n")
    f.close()
```

```python
from collections import defaultdict
# helper class to calculate COM (weighted) for specified atoms and a residue
def to_numpy(vec):
    return np.array([vec.x, vec.y, vec.z])
    
#helper class to calculate COM (weighted) for specified atoms and a residue
def com(residue, indices):
    #weight = defaultdict(int)
    weighted_vec = sum([(to_numpy(residue.xyz(x)) * weight[residue.atom_name(x)]) for x in indices])
    tot_mass = sum([weight[residue.atom_name(x)] for x in indices])
    return weighted_vec / tot_mass
```
